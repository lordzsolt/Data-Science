```python
from ipywidgets import IntProgress
from IPython.display import display


#to avoid code duplication we import the previously defined functions
#note that this *runs* the whole module
%run processFeatures.ipynb

outPath = "../testing_set/ocr_output/"
for (dirpath, dirnames, filenames) in os.walk(outPath):
    break

if '.DS_Store' in filenames :
    filenames.remove('.DS_Store')
```

```python
for filename in filenames:
    text = readFile(filename)
    print(filename)
    print("can borrow : " + canBorrow(text))
    print("min directors : " + str(findMinDirectors(text)))
    print("max directors : " + str(findMaxDirectors(text)))
    print("can directors borrow : " + canDirectorsBorrow(text))
    print("director resolution needed : " + resolutionNeeded(text))
    print("quorum : " + findQuorum(text))
    print("\n")
```

```python
resultsPath = "../results/"
with open(resultsPath + "results_testing.csv", "w") as text_file:
    text_file.write("companyName,canBorrow,minDirectors,maxDirectors,directorsBorrow,resolutionNeeded,quorum\n")
    progress = IntProgress(min=0, max=len(filenames)) # instantiate the progress bar
    display(progress) # display the progress bar
    for filename in filenames:
        text = readFile(filename)
        text_file.write("\"" + filename[:-4] + "\"" \
                       "," + canBorrow(text) + \
                       "," + str(findMinDirectors(text)) + \
                       "," + str(findMaxDirectors(text)) + \
                       "," + canDirectorsBorrow(text) + \
                       "," + resolutionNeeded(text) + \
                       "," + findQuorum(text) + "\n"
                       )
        progress.value += 1
        
```
